2. Find the total number of customers from each country. Display the country name and the corresponding count. Order the results by the count in descending order.

3. Identify the top 5 genres with the highest number of tracks. Display the genre name along with the total number of tracks for each genre.
   > Operations: `SELECT`, `COUNT`, `GROUP BY`, `ORDER BY`
4. Determine the average invoice total for each customer, considering both the album and individual track purchases. Display the customer's first and last name along with the average invoice total. Order the results by the average invoice total in descending order.
   > Operations: `SELECT`, `AVG`, `JOIN`, `GROUP BY`, `ORDER BY`
5. Identify the customer who spent the most on music purchases. Display the customer's first and last name, along with the total amount spent.

```python
import pandas as pd
import sqlite3

%load_ext sql
%config SqlMagic.autopandas = True
%config SqlMagic.displaycon = False

conn = sqlite3.connect("chinook.sqlite")

%sql sqlite:///chinook.sqlite
```

```python
#problem 1
result1 = %sql SELECT name, albumid FROM tracks LIMIT 5;
result1 = pd.DataFrame(data=result1)
result1.columns = ['track name','album title']
display(result1)
```

```python
#problem 2
result2 = %sql SELECT country , COUNT(*) FROM customers \
         GROUP BY country ORDER BY COUNT(*) DESC
result2 = pd.DataFrame(data=result2)
result2.columns = ['country','number of customers']
display(result2)
```

```python
#problem 3
# Query
result3 = %sql SELECT g.name, COUNT(t.trackid) FROM genres g \
         JOIN tracks t ON g.genreid = t.genreid \
         GROUP BY g.name ORDER BY COUNT(t.trackid) DESC LIMIT 5
result3 = pd.DataFrame(data=result3)
result3.columns = ['genre','number of tracks']
display(result3)
```

```python
#problem 4
result4 = %sql SELECT c.firstname, c.lastname \
        , AVG(i.total) FROM customers c \
        JOIN invoices i ON c.customerid = i.customerid \
        GROUP BY c.customerid ORDER BY AVG(i.total) DESC
result4 = pd.DataFrame(data=result4)
result4.columns = ['first name','last name','average invoice total']
display(result4)
```

```python
#problem 5
result5 = %sql SELECT c.firstname, c.lastname \
, SUM(i.total) FROM customers c \
JOIN invoices i ON c.customerid = i.customerid \
GROUP BY c.customerid ORDER BY SUM(i.total) DESC \
LIMIT 1

result5 = pd.DataFrame(data=result5)
result5.columns = ['first name','last name','total amount spent']
result5
```
